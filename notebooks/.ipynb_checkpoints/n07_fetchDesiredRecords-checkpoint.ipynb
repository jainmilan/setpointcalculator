```python
# Import Libraries
import os
import glob
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Read CSV and set date as the index column. Parsing date in a format to avoid confusion and speed up parsing.
df = pd.read_csv('..\\..\\data\\processed\\weather\\nrel\\nycNRELwSR.csv', index_col=[0], \
                 parse_dates=True, infer_datetime_format="%m/%d/%Y %H:%M:%S")
df.head()
```

```python
# Print columns of the dataframe 
df.columns
```

```python
SR_max = df[['SR_N', 'SR_S', 'SR_E', 'SR_W', \
             'SR_NE', 'SR_SE', 'SR_NW', 'SR_SW']].max().max()
SR_min = df[['SR_N', 'SR_S', 'SR_E', 'SR_W', \
             'SR_NE', 'SR_SE', 'SR_NW', 'SR_SW']].min().min()
SR_normalise_param = SR_max - SR_min
df[['SRP_N', 'SRP_S', 'SRP_E', 'SRP_W', \
    'SRP_NE', 'SRP_SE', 'SRP_NW', 'SRP_SW']] = (df[['SR_N', 'SR_S', 'SR_E', \
                                                    'SR_W', 'SR_NE', 'SR_SE', \
                                                    'SR_NW', 'SR_SW']] * 100.0) / SR_normalise_param

# Recommended Level (1000 lux): conversion to w/m2 is approximation based on literature, not very accurate
df['SR_REC'] = (1000 * 0.0079 * 100.0) / SR_normalise_param 
df.head()
```

```python
desired_cols = ['Dew Point', 'Relative Humidity', 'Temperature', 'SR_N', 'SR_S', \
                'SR_E', 'SR_W', 'SR_NE', 'SR_SE', 'SR_NW', 'SR_SW', 'SRP_N', 'SRP_S', \
                'SRP_E', 'SRP_W', 'SRP_NE', 'SRP_SE', 'SRP_NW', 'SRP_SW', 'SR_REC']
dfDesired = df[desired_cols].copy()
dfDesired.head()
```

```python
dfMedian = dfDesired.groupby([dfDesired.index.month, \
                              dfDesired.index.hour]).median().rename_axis(index=["Month", \
                                                                                 "Hour"])
dfMedian.columns = ['WBT', 'RH', 'DBT', 'SR_N', 'SR_S', 'SR_E', \
                    'SR_W', 'SR_NE', 'SR_SE', 'SR_NW', 'SR_SW', \
                    'SRP_N', 'SRP_S', 'SRP_E', 'SRP_W', 'SRP_NE', \
                    'SRP_SE', 'SRP_NW', 'SRP_SW', 'SR_REC']
dfMedian = dfMedian.reset_index()
dfMedian.head()
```

```python
dfMedian["SPT_BC"] = 70
# Recommendations for Bachelors - Cold (Jan, Feb, Dec)
dfMedian.loc[dfMedian.Month.isin([1, 2, 12]), "SPT_BC"] = 71
# Recommendations for Bachelors - Cool (Mar, Nov)
dfMedian.loc[dfMedian.Month.isin([3, 11]), "SPT_BC"] = 73
# Recommendations for Bachelors - Warm (Apr, May, Oct)
dfMedian.loc[dfMedian.Month.isin([4, 5, 10]), "SPT_BC"] = 74
# Recommendations for Bachelors - Hot (Jun, Jul, Aug, Sep)
dfMedian.loc[dfMedian.Month.isin([6, 7, 8, 9]), "SPT_BC"] = 75

dfMedian["SPT_FwoK"] = 70
# Recommendations for Family w/o kids - Cold (Jan, Feb, Dec)
dfMedian.loc[dfMedian.Month.isin([1, 2, 12]), "SPT_FwoK"] = 66
# Recommendations for Family w/o kids - Cool (Mar, Nov)
dfMedian.loc[dfMedian.Month.isin([3, 11]), "SPT_FwoK"] = 68
# Recommendations for Family w/o kids - Warm (Apr, May, Oct)
dfMedian.loc[dfMedian.Month.isin([4, 5, 10]), "SPT_FwoK"] = 69
# Recommendations for Family w/o kids - Hot (Jun, Jul, Aug, Sep)
dfMedian.loc[dfMedian.Month.isin([6, 7, 8, 9]), "SPT_FwoK"] = 70

dfMedian["SPT_FwK"] = 70
# Recommendations for Family w kids - Cold (Jan, Feb, Dec)
dfMedian.loc[dfMedian.Month.isin([1, 2, 12]), "SPT_FwK"] = 62
# Recommendations for Family w kids - Cool (Mar, Nov)
dfMedian.loc[dfMedian.Month.isin([3, 11]), "SPT_FwK"] = 65
# Recommendations for Family w kids - Warm (Apr, May, Oct)
dfMedian.loc[dfMedian.Month.isin([4, 5, 10]), "SPT_FwK"] = 66
# Recommendations for Family w kids - Hot (Jun, Jul, Aug, Sep)
dfMedian.loc[dfMedian.Month.isin([6, 7, 8, 9]), "SPT_FwK"] = 67

dfMedian["SPT_SC"] = 70
# Recommendations for Senior Citizens - Cold (Jan, Feb, Dec)
dfMedian.loc[dfMedian.Month.isin([1, 2, 12]), "SPT_SC"] = 75
# Recommendations for Senior Citizens - Cool (Mar, Nov)
dfMedian.loc[dfMedian.Month.isin([3, 11]), "SPT_SC"] = 77
# Recommendations for Senior Citizens - Warm (Apr, May, Oct)
dfMedian.loc[dfMedian.Month.isin([4, 5, 10]), "SPT_SC"] = 77
# Recommendations for Senior Citizens - Hot (Jun, Jul, Aug, Sep)
dfMedian.loc[dfMedian.Month.isin([6, 7, 8, 9]), "SPT_SC"] = 78
```

```python
dfMedian["ToDP"] = 1.63
dfMedian.loc[dfMedian.Month.isin([6, 7, 8, 9]) & dfMedian.Hour.isin(range(8, 24)), "ToDP"] = 23.07
dfMedian.loc[dfMedian.Month.isin([1, 2, 3, 4, 5, 10, 11, 12]) & dfMedian.Hour.isin(range(8, 24)), "ToDP"] = 8.54
```

```python
dfMedian["SP_l250"] = 10.87
dfMedian["SP_g250"] = 10.87
dfMedian.loc[dfMedian.Month.isin([6, 7, 8, 9]), "SP_g250"] = 12.434
```
