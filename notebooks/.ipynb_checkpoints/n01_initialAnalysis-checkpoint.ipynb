```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('..\\..\\data\\weather\\nyc_20191206_20191220.csv', skiprows=11, delimiter=',')
```

```python
df.head()
```

```python
df.keys()
```

```python
newCols = ['Year', 'Month', 'Day', 'Hour', 'Minute',\
          'Temperature', 'RH', 'MSLP', 'TP_HR', 'TP_LR', \
           'SA_HR', 'SA_LR', 'TCC', 'HCC', 'MCC', 'LCC', \
           'SSD', 'SWR', 'WS_10', 'WD_10', 'WS_80', 'WD_80', \
          'WS_900', 'WD_900', 'WG']
df.columns = newCols
```

```python
# Update date-time in a particular format
dateTimeStr = list()
for indx, row in df.iterrows():
    dateTimeStr.append(str(int(row["Month"])) + "/" + str(int(row["Day"])) + "/" + str(int(row["Year"])) + " " + \
                str(int(row["Hour"])) + ":" + str(int(row["Minute"])))
```

```python
# set date-time index
df["DateTimeStr"] = pd.to_datetime(dateTimeStr)
df = df.set_index("DateTimeStr")
```

```python
# update data frame
df.head()
```

# Outside Temperature

```python
ax = df["Temperature"].plot()
ax.set_ylabel('Temperature (degF)')
ax.set_xlabel('Date')

plt.tight_layout()
```

```python
df_hourly = df["Temperature"].groupby(df.index.hour)

# hour of the day variations
mean_hourly_temp = df_hourly.apply(np.mean)
std_hourly_temp = df_hourly.apply(np.std)

ax = mean_hourly_temp.plot(yerr=std_hourly_temp)
ax.set_ylabel('Temperature (degF)')
ax.set_xlabel('Hour of the Day')

plt.tight_layout()
```

# Sunlight

```python
ax = df["SSD"].plot(ylim=(-1, 70))
ax.set_ylabel('Sunshine Duration (%)')
ax.set_xlabel('Date')

plt.tight_layout()
```

```python
df_hourly = df["SSD"].groupby(df.index.hour)

# hour of the day variations
mean_hourly_temp = df_hourly.apply(np.mean)
std_hourly_temp = df_hourly.apply(np.std)

ax = mean_hourly_temp.plot(yerr=std_hourly_temp, ylim=(-75, 70))
ax.set_ylabel('Sunshine Duration (%)')
ax.set_xlabel('Hour of the Day')

plt.tight_layout()
```
